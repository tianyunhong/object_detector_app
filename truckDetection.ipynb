```python
from imageai.Detection import ObjectDetection
import os
import numpy as np
import cv2

execution_path = os.getcwd()
path = "/content/testfolder/*.jpg"
path_2 ="/content/result folder/"
path_crop ="/content/image_crop/"

detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( os.path.join(execution_path , "yolov3.pt"))
detector.loadModel()
img_number = 1

import glob
for file in glob.glob(path):
  #detections = detector.detectObjectsFromImage(input_image=os.path.join(execution_path , file), output_image_path= (path_2 + str(img_number)+".jpg"), minimum_percentage_probability=30)
  detections = detector.detectObjectsFromImage(input_image=file, output_image_path= (path_2 + str(img_number)+".jpg"), minimum_percentage_probability=93)
  
  #for eachObject in detections:
    #print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    #print("--------------------------------")
  for eachObject in detections:
    if eachObject["name"] == "truck":
      y_1 = eachObject["box_points"][0]
      y_2 = eachObject["box_points"][2]
      x_1 = eachObject["box_points"][1]
      x_2 = eachObject["box_points"][3]
     
  image_crop = cv2.imread(file)
  image_crop = image_crop[x_1:x_2,y_1:y_2]
  cv2.imwrite(path_crop +str(img_number)+"_crop.jpg",image_crop)
  img_number += 1

```

```python
image = crop_height_half(image)
```

```python
# Convert the images into a grayscale image
grayf = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
imgf = cv2.medianBlur(grayf, 5)
```

```python

# equalizing the contrast of a given grayscale image
imgf = cv2.equalizeHist(grayf)
# Contrast Limited Adaptive Histogram Equalization
clahef = cv2.createCLAHE(clipLimit=40)
img_eqHistf = clahef.apply(imgf)
# Thresholding for image segmentation (Otsu's Binarization)
threshf = cv2.threshold(img_eqHistf, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
circlesf = cv2.HoughCircles(threshf, cv2.HOUGH_GRADIENT, 1, 150, param1=200, param2=150, minRadius=30, maxRadius=75)
```

```python
import numpy as np
```

```python
# Thresholding for image segmentation (Otsu's Binarization)
thresh = cv2.threshold(img_eqHistf, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
circles = cv2.HoughCircles(thresh, cv2.HOUGH_GRADIENT, 1, 150, param1=200, param2=700, minRadius=30, maxRadius=75)

if circles is not None:
    # convert the (x, y) coordinates and radius of the circles to integers
    circles = np.round(circles[0, :]).astype("int")
    # draw the circle in the output image
    for (x, y, r) in circles:
        cv2.circle(image, (x, y), r, (36, 255, 12), 3)
```
